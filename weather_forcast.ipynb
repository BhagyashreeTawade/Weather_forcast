```python
#Monthly Average, Minimum, and Maximum Temperatures
# Ensure 'Date/Time' is in datetime format
df['Date/Time'] = pd.to_datetime(df['Date/Time'])

# Group by Month (directly without adding new columns)
# .dt.to_period("M")= converts datetime value into months

monthly_temp_stats = (
    df.groupby(df['Date/Time'].dt.to_period('M'))['Temp_C']
    .agg(['mean', 'min', 'max'])
    .reset_index()
)

# Rename columns for clarity
monthly_temp_stats.columns = ['Month', 'Average_Temp', 'Min_Temp', 'Max_Temp']

plt.figure(figsize=(12, 6))
plt.plot(monthly_temp_stats['Month'].astype(str), monthly_temp_stats['Average_Temp'],label='Average Temp',color='blue')
plt.plot(monthly_temp_stats['Month'].astype(str),monthly_temp_stats['Min_Temp'],label='Min Temp',color='green')
plt.plot(monthly_temp_stats['Month'].astype(str),monthly_temp_stats['Max_Temp'],label='Max Temp',color='red')
plt.title('Monthly Average, Minimum, and Maximum Temperatures')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

```

```python
corelation = df["Temp_C"].corr(df["Rel Hum_%"])
print(f"corelation between Temp_C and Rel Hum_% is {corelation}")
```

```python
#Avg visibility for each weather conditions
visibility_by_weather = df.groupby("Weather_Conditions")["Visibility_km"].mean()
visibility_by_weather
```

```python
df["Date/Time"] = pd.to_datetime(df["Date/Time"])
df["Day/Night"] = df["Date/Time"].apply(lambda x: 'Night' if 18 <= x.hour or x.hour < 6 else 'Day')
```

```python
avg_temp_day_night = df.groupby("Day/Night")["Temp_C"].mean()
avg_temp_day_night
```

```python
avg_temp_day_night.plot(kind = "bar", color=['skyblue', 'salmon'])
plt.title('Average Temperature During Day vs Night')
plt.xlabel('Day/Night')
plt.ylabel('Average Temperature (°C)')
plt.show()
```

```python
extreme_conditions = df[(df["Visibility_km"] < 5) | (df["Temp_C"] < 0)]
extreme_conditions
```

```python
#Temperature vs Dew Point Temperature
plt.figure(figsize=(8, 5))
plt.hexbin(df['Temp_C'], df['Dew Point Temp_C'], gridsize=30, cmap='Blues')
plt.colorbar(label='Density')
plt.xlabel('Temperature (°C)')
plt.ylabel('Dew Point Temperature (°C)')
plt.title('Hexbin Plot of Temperature vs Dew Point Temperature')
plt.show()
```
